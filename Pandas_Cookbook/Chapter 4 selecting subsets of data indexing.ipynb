## Selecting Series data

The indexing terminology can get confusing. The term indexing operator is used here to distinguish it from 
the other indexers. It refers to the brackets, []

> For instance, given a Series s, you can select data in the following ways: ***s[item]*** and ***s.loc[item].*** The first uses the indexing operator. The second uses the ***.loc*** indexer

````python
url ='https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/college.csv'
college = pd.read_csv(url, index_col='INSTNM')
city = college['CITY']
````

````python
college.head()
````

````python
city.head()
````

````python
#The .iloc indexer makes selections only by integer location.
city.iloc[3]
````

````python
#To select several different integer locations, pass a list to .iloc. (also, two [] shows the index)
city.iloc[[10,20,30]]
````

````python
#To select an equally spaced partition of data
city.iloc[4:50:10]
````

````python
#Now we turn to the .loc indexer, which selects only with index labels
city.loc['Heritage Christian University']
````

````python
#also
city.loc[['Heritage Christian University']]
````

````python
#To select an equally spaced partition of data, use slice notation. Make sure that the start and stop values are strings. 
#You can use an integer to specify the step size of the slice

#*********iloc and loc retunr rows*******
````

````python
city.loc['Alabama State University':'Reid State Technical College':10]
````

## Selecting DataFrame rows

```The most explicit and preferred way to select DataFrame rows is with the .iloc and .loc indexers.```

````python
college.head()
````

````python
#Pass an integer to the .iloc indexer to select an entire ***row*** at that position
college.iloc[60]
````

````python
#the same row wit loc would be
college.loc['University of Alaska Anchorage']
````

````python
# To select a disjointed set of rows as a DataFrame, pass a list of integers to the .iloc indexer:
college.iloc[[60, 99, 3]]
````

````python
# now with loc
labels = ['University of Alaska Anchorage', 'International Academy of Hair Design', 'University of Alabama in Huntsville']
college.loc[labels]
````

## Selecting DataFrame rows and columns simultaneously

To select rows and columns simultaneously, you will need to pass both valid row and column selections separated by a comma to either the .iloc or .loc indexers.

````python
college.iloc[:3, :4]
````

````python
college.loc[:'Amridge University',:'MENONLY']
````

## Selecting data with both integers and labels

The .iloc and .loc indexers each select data by either integer or label location but are not able to handle a
combination of both input types at the same time.

The following recipe shows how to find the integer location of the columns and then use .iloc to complete the selection.

````python
col_start= college.columns.get_loc('UGDS_WHITE')
col_end = college.columns.get_loc('UGDS_UNKN') +1 #el 1 se agrega para incluir esta columna
````

````python
college.iloc[:5, col_start:col_end]
````

````python
#podemos hacer los mismo con loc

row_start = college.index[10]
row_end = college.index[15]

college.loc[row_start:row_end, 'UGDS_WHITE':'UGDS_UNKN']
````

## Slicing rows lazily

This is just a shortcut to show additional features of pandas, but the primary function of the indexing operator 
is actually to select DataFrame columns.

If you want to select rows, it is best to use .iloc or .loc, as they are unambiguous.

````python
college[10:20:2]
````

````python
#the same with lazy index
city = college['CITY']
city[10:20:2]
````
