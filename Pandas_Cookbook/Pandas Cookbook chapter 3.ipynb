## Calculating a trailing stop order price

```One basic type of trade that many investors employ is the stop order.``` 

> A stop order is an order placed by an investor to buy or sell a stock that executes whenever the market price reaches a certain point.

````python
import pandas_datareader as pdr
````

````python
tsla = pdr.DataReader('tsla', data_source='yahoo', start='2017-1-1')
````

````python
tsla
````

````python
tsla_close = tsla['Close']
tsla_close
````

````python
# Use the cummax method to track the highest closing price until the current date
# The cummax method works by retaining the maximum value encountered up to and including the current value.
tsla_cummax = tsla_close.cummax()
tsla_cummax
````

````python
# To limit the downside to 10%, we multiply tsla_cummax by 0.9. This creates the trailing stop order
tsla_trailing_stop = tsla_cummax* .9
tsla_trailing_stop
````

````python
#a function for loss. It works in a way that if a price falls below the 15% of the highest price, we can see it and act
def set_trailing_loss(symbol, purchase_date, perc):
    close = pdr.DataReader(symbol, 'yahoo', start=purchase_date)['Close']
    return close.cummax() * perc
````

````python
msft_trailing_stop = set_trailing_loss('msft', '2017-6-1', .85)
msft_trailing_stop.head()
````
